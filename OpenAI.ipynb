**Hugging Face models in LangChain**

There are thousands of models freely available to download and use on Hugging Face. [Hugging Face](https://huggingface.co/models) integrates really nicely into LangChain via its partner library, langchain-huggingface.

```python
# For LangChain Hugging Face integration
#!pip install -q -U langchain-huggingface

# Import the HuggingFacePipeline class for defining Hugging Face pipelines
from langchain_huggingface import HuggingFacePipeline

# Define the LLM from the Hugging Face model ID
llm = HuggingFacePipeline.from_model_id(
    model_id="crumb/nano-mistral",
    task="text-generation",
    pipeline_kwargs={"max_new_tokens": 20}
)

prompt = "Hugging Face is"

# Invoke the model
response = llm.invoke(prompt)
print(response)
```

This syntax for defining and invoking LLMs unlocks the ability to use almost any model. Let's look at strategies for prompting these models effectively.

**Prompt templates and chaining**

We wukk be using two of the core components in LangChain: prompt templates and chains

```python
#Create a prompt template from the template string
template = 'You are an artificial intelligent assistant, answer the question {question}'
prompt = PromptTemplate.from_template(template=template)

llm= ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=API_KEY)

#Create a chain to integrate the prompt template and llm
llm_chain = prompt | llm

#Invoke the chain on the question
question = 'How does LangChain make application development easy?'
response = llm_chain.invoke({'question': question})

print(response.content)
```

We perfectly pieced-together prompt produced pristine performance! Well-designed prompt templates are at the heart of many LangChain applications; keep experimenting to find the template that gives us optimum performance

**Chat prompt templates**

Given the importance of chat models in many LLM applications, LangChain provides functionality for creating prompt templates to structure messages to different chat roles.

```python
llm= ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=API_KEY)

#Create a chat prompt template
prompt_template_flag = ChatPromptTemplate.from_messages(
    [
        ('system', 'You are a geography expert that returns the colors present in the countries flag'),
        ('human', 'France'),
        ('ai', 'blue, white, red'),
        ("human", '{country}')
    ]
)

#Chain the prompt template and the model and invoke the chain
llm_chain = prompt_template_flag | llm
response = llm_chain.invoke({'country': 'India'})
print(response.content)
```

This is the power of prompt templates for translating user inputs into model inputs.

**Creating the few-shot example set**

PromptTemplate and ChatPromptTemplate are great for integrating variables, but struggle with integrating datasets containing many examples. This is where FewShotPromptTemplate comes in

```python
#Create the example list of dicts
examples = [
    {'question':'Who is the CEO of Facebook?',
      'answer':'Mark Zuckerberg'},
    {'question':'Who is the CEO of Apple?',
      'answer':'Tim Cook'},
    {'question':'Who is the CEO of Google?',
     'answer':'Sundar Pichai'}

]

#Convert pandas dataframe to a list of dict
#examples = df.to_dict(orient='records')
```

```python
#Complete the prompt for formatting answers
example_prompt = PromptTemplate.from_template("Question: {question}\nAnswer: {answer}")

#Create the few shot prompt
prompt_template = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    suffix = "Question: {input}\nAnswer:",
    input_variables=['input']
)

prompt = prompt_template.invoke({"input": "The name of Apple CEO is?"})
print(prompt.text)
```

Invoking the prompt template allows you to see exactly what context the model will have

```python
#Create a Gemini chat LLM
llm= ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=API_KEY)

#Create and invoke the chain
llm_chain = prompt_template|llm
response = llm_chain.invoke({"input": "The name of Apple CEO is?"})
print(response.content)
```

 Being able to integrate external data into prompts is an incredibly valuable skill for designing LLM applications like chatbots.
